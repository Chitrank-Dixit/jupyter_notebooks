```python
# Singly linked lists
# Node of the singly linked list

class Node:
    def __init__(self):
        self.data = None
        self.next_ptr = None
        self.length = 0
        self.head = None
        
    def set_data(self, data):
        self.data = data
        
    def get_data(self):
        return self.data
    
    def set_next(self, next_ptr):
        self.next_ptr = next_ptr
        
    def get_next(self):
        return self.next_ptr
    
    def has_next(self):
        return self.next_ptr != None
    
            
class LinkedList:

    def __init__(self, head=None, length=0):
        self.head = head
        self.length = 0
        
    def list_length(self):
        current = self.head
        count = 0
        while current != None:
            count += 1
            current = current.get_next()
        return count
    
    def get_all_elements_of_linked_list(self):
        current = self.head
        linked_list = []
        while current != None:
            linked_list.append((current.get_data() , current.get_next()))
            current = current.get_next()
        return linked_list
    
    # searching the element in the singly linked list
    def search(self, data, no_of_items=1):
        current = self.head
        found_list = []
        while current and no_of_items !=0:
            if current.get_data() == data:
                no_of_items -= 1
                found_list.append(current)
                current = current.get_next()
            else:
                current = current.get_next()
        if current is None:
            raise ValueError("Data not in list")
        return found_list
    
    # inserting an element before the head of the linked list

    def insert_at_beginning(self, data):
        new_node = Node()
        new_node.set_data(data)

        if self.length == 0:
            self.head = new_node
        else:
            new_node.set_next(self.head)
            self.head = new_node
            
        self.length += 1
        
    # inserting the node at the end of the linked list
    def insert_at_end(self, data):
        new_node = Node()
        new_node.set_data(data)
        
        current = self.head

        while current.get_next() != None:
            current = current.get_next()
            
        current.set_next(new_node)
        self.length += 1
        
    # inserting element in the middle of the linked list
    
    def insert_at_position(self, position, data):
        if position > self.length or position < 0:
            return None
        else:
            if position == 0:
                self.insert_at_beginning(data)
            else:
                new_node = Node()
                new_node.set_data(data)
                count = 0
                current = self.head
                while count < position -1:
                    count += 1
                    current = current.get_next()
                
                new_node.set_next(current.get_next())
                current.set_next(new_node)
                self.length += 1
                
    # deleting an element from the beginning
    def delete_from_beginning(self):
        if self.length == 0:
            print "list is empty"
        else:
            self.head = self.head.get_next()
            self.length -= 1
            
    # deleting the last element of the linked list
    def delete_from_end(self):
        if self.length == 0:
            print "list is empty"
        else:
            current_node = self.head
            previous_node = self.head
            while current_node.get_next() != None:
                previous_node = current_node
                current_node = current_node.get_next()
                
            previous_node.set_next(None)
            self.length -= 1
            
            
    # deleting a node from any position of the node in linked list
    def delete_from_node(self, node):
        if self.length == 0:
            print "list is empty"
        else:
            current = self.head
            previous = None
            found = False
            while not found:
                if current == node:
                    found = True
                elif current is None:
                    raise ValueError("Node not in Linked List")
                else:
                    previous = current
                    current = current.get_next()
        if previous is None:
            self.head = current.get_next()
        else:
            previous.set_next(current.get_next())
            
        self.length -= 1
        
    # delete with value from the linked list
    def delete_from_data(self, data):
        if self.length == 0:
            print "list is empty"
        else:
            current_node = self.head
            previous_node = self.head
            
            while current_node.next_ptr != None or current_node.data != data:
                if current_node.data == data:
                    previous_node.next_ptr = current_node.next_ptr
                    self.length -= 1
                    return 
                else:
                    previous_node = current_node
                    current_node = current_node.next_ptr
                    
            print "The data provided is not present"
            
    # delete a node by position of that node
    def delete_from_position(self, position):
        if self.length == 0:
            print "list is empty"
        else:
            count = 0
            current_node = self.head
            previous_node = self.head
            
            if position > self.length or position < 0:
                print "The position does not exist. Please enter a valid position"
            else:
                while current_node.next_ptr != None or count < position:
                    count += 1
                    if count == position:
                        previous_node.next_ptr = current_node.next_ptr
                        self.length -= 1
                        return
                    else:
                        previous_node = current_node
                        current_node = current_node.next_ptr
    
    def clear(self):
        self.head = None

ll = LinkedList()

# insert at the beginning of the linked list
ll.insert_at_beginning(12)
ll.insert_at_beginning(75)
ll.insert_at_beginning(90)
print ll.get_all_elements_of_linked_list()

# insert at the beggining of the linked list
ll.insert_at_beginning(23)

print ll.get_all_elements_of_linked_list()

# get number of elements currently in the linked list
print ll.list_length()

# insert element at the end of the linked list
ll.insert_at_end(37)
ll.insert_at_end(37)
ll.insert_at_end(46)
ll.insert_at_end(31)
ll.insert_at_end(67)
ll.insert_at_end(89)

# get all element of the linked list
print ll.get_all_elements_of_linked_list()

# this would use to find the 2 occurances of linked list element 37
print ll.search(37, 2)


# insert at specific position in the linked list
# position works with 0, 1, 2 etc
ll.insert_at_position(2, 45)
print ll.get_all_elements_of_linked_list()


# deleting an element from the beginning
ll.delete_from_beginning()
print ll.get_all_elements_of_linked_list()


# deleting an element from the end
ll.delete_from_end()
print ll.get_all_elements_of_linked_list()

# delete by specifying data
ll.delete_from_data(45)
print ll.get_all_elements_of_linked_list()

# delete by position
ll.delete_from_position(2)
print ll.get_all_elements_of_linked_list()

# deleting an element from the node 
# This takes address of the node and deletes that node at that address
nodes = ll.get_all_elements_of_linked_list()
ll.delete_from_node(nodes[-2][1])
print ll.get_all_elements_of_linked_list()

# delete all elements from linked list
ll.clear()
print ll.get_all_elements_of_linked_list()
```

```python
# doubly linked list

class Node:
    def __init__(self, data=None, next_ptr=None, prev_ptr=None):
        self.data = data
        self.next_ptr = next_ptr
        self.length = 0
        self.prev_ptr = prev_ptr
        #self.head = None
        
    def set_data(self, data):
        self.data = data
        
    def get_data(self):
        return self.data
    
    def set_next(self, next_ptr):
        self.next_ptr = next_ptr
        
    def get_next(self):
        return self.next_ptr
    
    def has_next(self):
        return self.next_ptr != None
    
    def set_prev(self, prev_ptr):
        self.prev_ptr = prev_ptr
        
    def get_prev(self):
        return self.prev_ptr
    
    def has_prev(self):
        return self.prev_ptr != None
    
    def __str__(self):
        return "Node (Data = %s)" % (self.data)
    
    
class DoublyLinkedList:

    def __init__(self, head=None, tail=None, length=0):
        self.head = head
        self.tail = tail
        self.length = 0
        
    def list_length(self):
        current = self.head
        count = 0
        while current != None:
            count += 1
            current = current.get_next()
        return count
    
    def get_all_elements_of_linked_list(self):
        current = self.head
        linked_list = []
        while current != None:
            linked_list.append((current.get_prev(), current.get_data() , current.get_next()))
            current = current.get_next()
        return linked_list
    
    # searching the element in the singly linked list
    def search(self, data, no_of_items=1):
        current = self.head
        found_list = []
        while current and no_of_items !=0:
            if current.get_data() == data:
                no_of_items -= 1
                found_list.append(current)
                current = current.get_next()
            else:
                current = current.get_next()
        if current is None:
            raise ValueError("Data not in list")
        return found_list
    
    # inserting an element before the head of the linked list

    def insert_at_beginning(self, data):
        new_node = Node(data, None, None)
        if self.head == None:
            self.head = self.tail = new_node
        else:
            new_node.set_prev(None)
            new_node.set_next(self.head)
            self.head.set_prev(new_node)
            self.head = new_node    
            self.length += 1
        
    # inserting the node at the end of the linked list
    def insert_at_end(self, data):
        #new_node = Node(data, None, None)
        if self.head == None:
            self.head = Node(data, None, None)
            self.tail = self.head
        else:
            current = self.head
            while (current.get_next() != None):
                current = current.get_next()
            current.set_next(Node(data, None, current))
            self.tail = current.get_next()  
            self.length += 1
        
    # inserting element in the middle of the linked list
    def get_node(self, index):
        current_node = self.head
        if current_node == None:
            return None
        i=1
        while i<= index:
            current_node = current_node.get_next()
            if current_node == None:
                break
            i += 1
        return current_node
    
    
    def insert_at_position(self, index, data):
        new_node = Node(data)
        if self.head == None or index == 0:
            self.insert_at_beginning(data)
        elif index > 0:
            temp = self.get_node(index-1)
            if temp == None or temp.get_next() == None:
                self.insert_at_end(data)
            else:
                new_node.set_next(temp.get_next())
                new_node.set_prev(temp.get_prev().get_next())
                temp.get_next().set_prev(new_node)
                temp.set_next(new_node)

        self.length += 1
                
    # deleting an element from the beginning
    def delete_from_beginning(self):
        if self.length == 0:
            print "list is empty"
        else:
            self.head = self.head.get_next()
            self.head.set_prev(None)
            self.length -= 1
            
    # deleting the last element of the linked list
    def delete_from_end(self):
        if self.length == 0:
            print "list is empty"
        else:
            current_node = self.head
            previous_node = self.head
            while current_node.get_next() != None:
                previous_node = current_node
                current_node = current_node.get_next()
                
            previous_node.set_next(None)
            self.length -= 1
            
            
    # deleting a node from any position of the node in linked list
    def delete_from_node(self, node):
        if self.length == 0:
            print "list is empty"
        else:
            current = self.head
            previous = None
            found = False
            while not found:
                if current.get_data() == node:
                    found = True
                    current.get_prev().set_next(current.get_next())
                    current.get_next().set_prev(current.get_prev())
                    current.set_prev(None)
                    current.set_next(None)
                    current.set_data(None)
                elif current is None:
                    raise ValueError("Node not in Linked List")
                else:
                    previous = current
                    current = current.get_next()
        self.length -= 1
        
    # delete with value from the linked list
    
    
    def delete_from_data(self, data):
        temp = self.head
        while temp is not None:
            if temp.get_data() == data:
                if temp.get_next() is not None:
                    temp.get_prev().set_next(temp.get_next())
                    temp.get_next().set_prev(temp.get_prev())
                    temp.set_prev(None)
                    temp.set_next(None)
                    temp.set_data(None)
                else:
                    self.head = temp.get_next()
                    temp.get_next().set_prev(None)
                self.length -= 1
            temp = temp.get_next()
            
    # delete a node by position of that node
    def delete_from_position(self, index):
        temp = self.get_node(index - 1)
        if temp:
            if temp.get_next():
                temp.get_prev().set_next(temp.get_next())
                temp.get_next().set_prev(temp.get_prev())
                temp.set_prev(None)
                temp.set_next(None)
                temp.set_data(None)
            self.length -= 1
    
    def clear(self):
        self.head = None
        self.tail = None

    
dll = DoublyLinkedList()

# insert at the beginning of the linked list
dll.insert_at_beginning(12)
dll.insert_at_beginning(75)
dll.insert_at_beginning(90)
print dll.list_length()
print dll.get_all_elements_of_linked_list()

# insert at the beggining of the linked list
dll.insert_at_beginning(23)

print dll.get_all_elements_of_linked_list()

# get number of elements currently in the linked list
print dll.list_length()

# insert element at the end of the linked list
dll.insert_at_end(37)
dll.insert_at_end(37)
dll.insert_at_end(46)
dll.insert_at_end(31)
dll.insert_at_end(67)
dll.insert_at_end(89)

# get all element of the linked list
print dll.get_all_elements_of_linked_list()
print dll.list_length()
# # this would use to find the 2 occurances of linked list element 37
print dll.search(37, 2)


# # insert at specific position in the linked list
# # position works with 0, 1, 2 etc
dll.insert_at_position(2, 45)
print dll.get_all_elements_of_linked_list()
print dll.list_length()

# # deleting an element from the beginning
dll.delete_from_beginning()
print dll.get_all_elements_of_linked_list()


# deleting an element from the end
dll.delete_from_end()
print dll.get_all_elements_of_linked_list()

# TODO: need to make it work
# delete by specifying data
dll.delete_from_data(45)
print dll.get_all_elements_of_linked_list()

# # delete by position
dll.delete_from_position(2)
print dll.get_all_elements_of_linked_list()
#
# # TODO: need to make it work
# # # deleting an element from the node
# # # This takes address of the node and deletes that node at that address
nodes = dll.get_all_elements_of_linked_list()
dll.delete_from_node(nodes[-2][1])
print dll.get_all_elements_of_linked_list()
#
# # # delete all elements from linked list
dll.clear()
print dll.get_all_elements_of_linked_list()

```
