```python
# linked list implementation of Stack

class Node:
    def __init__(self):
        self.data = None
        self.next_ptr = None
        self.length = 0
        self.head = None
        
    def set_data(self, data):
        self.data = data
        
    def get_data(self):
        return self.data
    
    def set_next(self, next_ptr):
        self.next_ptr = next_ptr
        
    def get_next(self):
        return self.next_ptr
    
    def has_next(self):
        return self.next_ptr != None
    

class Stack(object):
    def __init__(self, data=None):
        self.head = None
        if data:
            for element in data:
                self.push(element)
                
    def get_all_elements_of_stack(self):
        current = self.head
        stack = []
        while current != None:
            stack.append((current.get_data() , current.get_next()))
            current = current.get_next()
        return stack
                
    def push(self, data):
        temp = Node()
        temp.set_data(data)
        temp.set_next(self.head)
        self.head = temp
        
    def pop(self):
        if self.head is None:
            raise IndexError
        temp = self.head.get_data()
        self.head = self.head.get_next()
        return temp
    
    def peek(self):
        if self.head is None:
            raise IndexError
        return self.head.get_data()
    
    
our_list = [4,3,2,1]
our_stack = Stack(our_list)
our_stack.push(24)
print our_stack.get_all_elements_of_stack()
print our_stack.pop()
print our_stack.pop()
print our_stack.get_all_elements_of_stack()
print our_stack.peek()
        
    
```
